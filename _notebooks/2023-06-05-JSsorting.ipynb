# Selection Sort
with selection sort, you repeatedly find the minimum element from the unsorted part of the list and swap it with the element at the beginning of the unsorted part.

## Try it!
### Run the code below to see how selection sort works!

```python
%%html
<html>
<head>
  <title>Selection Sort</title>
  <script>
    function selectionSort() {
      var numbers = document.getElementById("numbers1").value.split(",");
      var output = document.getElementById("output1");

      // Convert strings to numbers
      numbers = numbers.map(Number);

      var n = numbers.length;

      // Iterate through each element of the list
      for (var i = 0; i < n - 1; i++) {
        var minIndex = i;

        // Find the index of the minimum element in the unsorted part of the list
        for (var j = i + 1; j < n; j++) {
          if (numbers[j] < numbers[minIndex]) {
            minIndex = j;
          }
        }

        // Swap the minimum element with the first unsorted element
        var temp = numbers[minIndex];
        numbers[minIndex] = numbers[i];
        numbers[i] = temp;

        // Print the current step
        output.innerHTML += "Step " + (i + 1) + ": " + numbers.join(", ") + "<br>";
      }

      // Print the final sorted list
      output.innerHTML += "Sorted List: " + numbers.join(", ") + "<br>";
    }
  </script>
</head>
<body>
  <h2>Selection Sort</h2>
  <label for="numbers1">Enter numbers (separated by commas):</label><br>
  <input type="text" id="numbers1" name="numbers1"><br>
  <button onclick="selectionSort()">Sort</button><br><br>
  <div id="output1"></div>
</body>
</html>
```

# Bubble Sort
with bubble sort, you compare the two elements in the list at a time and swap the order if it is wrong, the same step is repeated to the next two elements in the list until the list is completely sorted.

## Try it!
### Run the code below to see how bubble sort works!

```python
%%html
<html>
<head>
  <title>Bubble Sort</title>
  <script>
    function bubbleSort() {
      var numbers = document.getElementById("numbers2").value.split(",");
      var output = document.getElementById("output2");

      // Convert strings to numbers
      numbers = numbers.map(Number);

      var n = numbers.length;

      // Iterate through the list
      for (var i = 0; i < n - 1; i++) {
        // Perform a single pass through the list
        for (var j = 0; j < n - i - 1; j++) {
          // Swap adjacent elements if they are in the wrong order
          if (numbers[j] > numbers[j + 1]) {
            var temp = numbers[j];
            numbers[j] = numbers[j + 1];
            numbers[j + 1] = temp;
          }
        }

        // Print the current step
        output.innerHTML += "Step " + (i + 1) + ": " + numbers.join(", ") + "<br>";
      }

      // Print the final sorted list
      output.innerHTML += "Sorted List: " + numbers.join(", ") + "<br>";
    }
  </script>
</head>
<body>
  <h2>Bubble Sort</h2>
  <label for="numbers2">Enter numbers (separated by commas):</label><br>
  <input type="text" id="numbers2" name="numbers2"><br>
  <button onclick="bubbleSort()">Sort</button><br><br>
  <div id="output2"></div>
</body>
</html>
```

# Merge Sort
with merge sort you divide the list into two until no longer divisible (creating a triangle), then you merge the elements back together in order from least to greatest until the full list is put back together (creating a diamond)

## Try it!
### Run the code below to see how merge sort works!

```python
%%html
<html>
<head>
  <title>Merge Sort</title>
  <script>
    function mergeSort(arr) {
      if (arr.length <= 1) {
        return arr;
      }

      var mid = Math.floor(arr.length / 2);
      var left = arr.slice(0, mid);
      var right = arr.slice(mid);

      left = mergeSort(left);
      right = mergeSort(right);

      return merge(left, right);
    }

    function merge(left, right) {
      var result = [];
      var i = 0;
      var j = 0;

      while (i < left.length && j < right.length) {
        if (left[i] < right[j]) {
          result.push(left[i]);
          i++;
        } else {
          result.push(right[j]);
          j++;
        }
      }

      while (i < left.length) {
        result.push(left[i]);
        i++;
      }

      while (j < right.length) {
        result.push(right[j]);
        j++;
      }

      return result;
    }

    function performMergeSort() {
      var numbers = document.getElementById("numbers5").value.split(",");
      var output = document.getElementById("output5");

      // Convert strings to numbers
      numbers = numbers.map(Number);

      var sortedNumbers = mergeSort(numbers);

      // Print the current step
      for (var i = 0; i < sortedNumbers.length; i++) {
        output.innerHTML += "Step " + (i + 1) + ": " + sortedNumbers.slice(0, i + 1).join(", ") + "<br>";
      }

      // Print the final sorted list
      output.innerHTML += "Sorted List: " + sortedNumbers.join(", ") + "<br>";
    }
  </script>
</head>
<body>
  <h2>Merge Sort</h2>
  <label for="numbers5">Enter numbers (separated by commas):</label><br>
  <input type="text" id="numbers5" name="numbers5"><br>
  <button onclick="performMergeSort()">Sort</button><br><br>
  <div id="output5"></div>
</body>
</html>
```

# Insertion Sort
with insertion sort you compare elements in pairs. Starting with the first pair you compare, if they are in ascending order then you leave it as is and move on to the next pair to the right. However, if it is not in ascending order, then the pair is swapped. If, down the line, a number that is smaller than all the previous ones occur, then it will continue to get swapped until it is pushed up to the front.

## Try it!
### Run the code below to see how insertion sort works!

```python
%%HTML
<html>
<head>
  <title>Insertion Sort</title>
  <script>
    function insertionSort() {
      var numbers = document.getElementById("numbers4").value.split(",");
      var output = document.getElementById("output4");

      // Convert strings to numbers
      numbers = numbers.map(Number);

      var n = numbers.length;

      // Iterate through the list
      for (var i = 1; i < n; i++) {
        var key = numbers[i];
        var j = i - 1;

        // Move elements of the sorted subarray that are greater than the key
        // to one position ahead of their current position
        while (j >= 0 && numbers[j] > key) {
          numbers[j + 1] = numbers[j];
          j--;
        }

        numbers[j + 1] = key;

        // Print the current step
        output.innerHTML += "Step " + i + ": " + numbers.join(", ") + "<br>";
      }

      // Print the final sorted list
      output.innerHTML += "Sorted List: " + numbers.join(", ") + "<br>";
    }
  </script>
</head>
<body>
  <h2>Insertion Sort</h2>
  <label for="numbers4">Enter numbers (separated by commas):</label><br>
  <input type="text" id="numbers4" name="numbers4"><br>
  <button onclick="insertionSort()">Sort</button><br><br>
  <div id="output4"></div>
</body>
</html>
```
